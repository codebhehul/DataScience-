# 1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a versatile library for data visualization. Here are five types of plots you can create with Seaborn and their common use cases:

1. Scatter Plot: Use sns.scatterplot() to display individual data points as dots. This is useful for visualizing the relationship between two continuous variables.

2. Bar Plot: Create bar charts using sns.barplot(). It's useful for comparing the values of different categories.

3. Box Plot: Use sns.boxplot() to visualize the distribution of a dataset and identify outliers. It's suitable for comparing multiple categories or distributions.

4. Histogram: Plot histograms with sns.histplot() or sns.distplot(). These are useful for understanding the distribution of a single variable.

5. Heatmap: Create heatmaps using sns.heatmap() to visualize the correlation or relationship between two-dimensional data.

# 2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.show()
```

# 3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.tight_layout()
plt.show()
```

# 4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True)
plt.show()
```

# 5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species")
plt.show()
```

# 6. Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.show()
```
