# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python used for creating static, interactive, and animated visualizations in various formats such as graphs, plots, charts, and more. It provides a wide range of functionalities for creating high-quality visualizations, making it a valuable tool for data analysts, scientists.

Matplotlib's Pyplot module is a sub-library that provides a simple interface for creating various types of plots quickly and easily. It is inspired by MATLAB's plotting capabilities and is commonly used for creating basic visualizations without requiring extensive code.

Five types of plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot: A simple plot that represents data points as connected lines, commonly used for showing trends over time or continuous data.
2. Bar Plot: Displays data as rectangular bars with heights or lengths proportional to the values they represent. Bar plots are effective for comparing categorical data.
3. Scatter Plot: A plot that displays individual data points as markers, useful for visualizing the relationship between two variables.
4. Histogram: Represents the distribution of a continuous dataset by dividing it into bins and showing the frequency of data points within each bin.
5. Pie Chart: Displays data as sectors of a circle, with each sector representing a proportion of the whole. Pie charts are suitable for showing parts-to-whole relationships.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot:- 

import numpy as np np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that uses points to represent individual data observations. It is used to display the relationship between two continuous variables, where one variable is plotted on the x-axis and the other on the y-axis. Scatter plots are particularly useful for identifying patterns, trends, clusters, and outliers in data.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot of X and Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. This is particularly useful when want to display multiple plots in a single figure to compare or visualize different aspects of data side by side. The function allows to specify the number of rows, the number of columns, and the index of the current subplot want to create.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of data visualization that uses rectangular bars to represent categorical data. It displays the values of different categories as bars with lengths or heights proportional to the values they represent. Bar plots are commonly used to compare and show the relationships between different categories or groups.

Bar plots are used for several purposes:
1. Comparison: Bar plots are effective for comparing the values of different categories. They make it easy to visually assess differences or similarities among categories.
2. Categorical Data: Bar plots are particularly useful for visualizing categorical data, where each category represents a distinct group or class.
3. Part-to-Whole Relationships: Bar plots can also be used to display part-to-whole relationships by showing the contributions of different components to a total value.
4. Trends: Bar plots can help visualize trends over time or across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='blue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='green')
plt.title('Company Profits')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a type of data visualization that provides a graphical summary of the distribution of a dataset. It displays key statistics such as the median, quartiles, and potential outliers, making it useful for understanding the central tendency, spread, and presence of extreme values in the data. Box plots are commonly used for comparing the distribution of data across different categories or groups.

A box plot consists of several components:
1. Box: The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data. The box spans the middle 50% of the data.
2. Median Line: A line inside the box represents the median (50th percentile) of the data.
3. Whiskers: Whiskers extend from the box to the minimum and maximum values within a certain range, typically 1.5 times the IQR. Data points outside this range are considered potential outliers and are plotted individually as points.
4. Outliers: Individual data points that are significantly different from the rest of the data and are plotted as points beyond the whiskers.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['Box 1', 'Box 2']
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=labels, vert=True, patch_artist=True)
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.show()
```
